```python
import pandas as pd
sales = pd.read_csv("DataSource/Sales.csv")
stores = pd.read_csv("DataSource/Stores.csv")
Products = pd.read_csv("DataSource/Products.csv")
customers = pd.read_csv("DataSource/Customers.csv",encoding='ISO-8859-1')
exchangerates = pd.read_csv("DataSource/Exchange_Rates.csv")
```

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'], format='%m/%d/%Y')
sales['Delivery Date'] = pd.to_datetime(sales['Delivery Date'], format='%m/%d/%Y', errors='coerce')
sales['Currency Code'] = sales['Currency Code'].astype(str)
```

```python
sales.fillna('Not Available', inplace=True)
sales
```

```python
merged_data = pd.merge(sales, customers, how='inner', on='CustomerKey')
```

```python
merged_data = pd.merge(merged_data, exchangerates, left_on=['Order Date', 'Currency Code'], right_on=['Date', 'Currency'])
```

```python
merged_data = pd.merge(merged_data, Products, on='ProductKey')
```

```python
merged_data = pd.merge(merged_data, stores, on=['StoreKey', 'Country', 'State'])
```
